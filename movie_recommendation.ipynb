```python
import numpy as np
import pandas as pd
```

```python
#import data set
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
credits
```

```python
#merge the two data
movies = movies.merge(credits,on='title')
```

```python
#columns to fetch - genres , id, keywords, title, overview, cast, crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#check for the missing data
movies.isnull().sum()
```

```python
#drop the data with the missing values
movies.dropna(inplace=True)
```

```python
#to check the duplicate data
movies.duplicated().sum()
```

```python
movies.drop_duplicates(inplace=True)
```

```python
movies.head()
```

```python
movies.iloc[0].genres
```

```python
import ast  # to convert string into integer
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
                 
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# get the first three cast from the cast_y
def convert_cast(obj):
    cast = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if counter != 3:
            cast.append(i['name'])
            counter += 1
        else:
             break
    return cast
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
#from the crew member get the director name
def director_cast(obj):
    cast = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            cast.append(i['name'])
            break
    return cast
```

```python
movies['crew'] = movies['crew'].apply(director_cast)
```

```python
movies.head()
```

```python
#convert the overview in the list
movies['overview'] = [overview.split() for overview in movies['overview']]
```

```python
movies.head()
```

```python
#remove the space so that model should not hallicunate
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#create the new column and concatenate oveview, genres, keywords, cast and crew

movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['keywords'] + movies['crew']
```

```python
movies.head()
```

```python
#create the new dataframe which will have movie_id, title and tags
new_movies_df = movies[['movie_id', 'title', 'tags']]
new_movies_df
```

```python
#now convert the tags into string
new_movies_df['tags'] = new_movies_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies_df.head()
```

```python
#convert tags into lower case
new_movies_df['tags'] = new_movies_df['tags'].apply(lambda x:x.lower())
```

```python
new_movies_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#helper function to stem the words
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
#apply stemming to convert for example the names or name or naming to name 
new_movies_df['tags'] = new_movies_df['tags'].apply(stem)
```

```python
#vectorization 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies_df['tags']).toarray()
```

```python
vectors[0]
```

```python
new_movies_df['tags'][0]
```

```python
#To find the similarity between the two vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
calculate_similarity = cosine_similarity(vectors)  #it will calculate the distance between the vectors
#cosine_similarity(vectors).shape
calculate_similarity
#use enumerate and create index position and convert into list , so that index won't get change when performing sorting
sorted(list(enumerate(calculate_similarity[0])),reverse = True, key = lambda x:x[1])[1:6]   

#to fetch the name of the movie
new_movies_df.iloc[11].title 
```

```python
#create the function to get the movie name and check for the similarity
def recommend_movie(movie):
    movie_index = new_movies_df[new_movies_df['title'] == movie].index[0]
    distance = calculate_similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse = True, key = lambda x:x[1])[1:6] 
    for i in movies_list:
        print(new_movies_df.iloc[i[0]].movie_id)
       
    
```

```python
recommend_movie('Batman Begins')
```
